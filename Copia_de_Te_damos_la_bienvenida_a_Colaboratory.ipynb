````
# -*- coding: utf-8 -*-
"""
AN√ÅLISIS DE MACHINE LEARNING EPIDEMIOL√ìGICO - C√ÅNCER DE PULM√ìN M√âXICO
======================================================================
Adaptado del sistema original de c√°ncer de mama para c√°ncer de pulm√≥n
Considerando las diferencias epidemiol√≥gicas espec√≠ficas:
- Mayor incidencia en hombres vs mujeres
- Rango etario diferente (t√≠picamente mayor)
- Factores de riesgo distintos (tabaquismo, exposici√≥n ocupacional)
"""

import os
import pickle
import time
import threading
import json
from datetime import datetime
import warnings
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix, silhouette_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB

warnings.filterwarnings('ignore')

print("ü´Å SISTEMA DE ML EPIDEMIOL√ìGICO - C√ÅNCER DE PULM√ìN M√âXICO")
print("=" * 70)
print("üìä Adaptado para an√°lisis espec√≠fico de c√°ncer pulmonar")
print("üë• Considerando mayor incidencia masculina y patrones etarios diferentes")

# ============================================================================
# PASO 1: SISTEMA DE PROTECCI√ìN (MANTENIDO IGUAL)
# ============================================================================

class SmartAutoSave:
    """Sistema inteligente de auto-guardado"""
    def __init__(self, interval_minutes=5):
        self.interval_seconds = interval_minutes * 60
        self.is_active = False
        self.thread = None
        self.data_registry = {}
        self.last_save = time.time()
        self.save_count = 0

    def register(self, name, data):
        """Registrar datos para auto-guardado"""
        self.data_registry[name] = {
            'data': data,
            'registered_at': datetime.now(),
            'size_mb': self._get_size_mb(data)
        }
        print(f"üìù Registrado para backup: {name} ({self._get_size_mb(data):.1f}MB)")

    def _get_size_mb(self, obj):
        """Estimar tama√±o del objeto en MB"""
        try:
            return sys.getsizeof(pickle.dumps(obj)) / (1024 * 1024)
        except:
            return 0.0

    def start(self):
        """Iniciar auto-guardado"""
        if not self.is_active:
            self.is_active = True
            self.thread = threading.Thread(target=self._save_loop, daemon=True)
            self.thread.start()
            print("üîÑ Auto-guardado INICIADO")

    def stop(self):
        """Detener auto-guardado"""
        self.is_active = False
        if self.thread:
            self.thread.join(timeout=5)
        print("‚è∏Ô∏è Auto-guardado DETENIDO")

    def _save_loop(self):
        """Bucle principal de guardado"""
        while self.is_active:
            try:
                time.sleep(30)
                if time.time() - self.last_save >= self.interval_seconds:
                    self._perform_save()
            except Exception as e:
                print(f"‚ö†Ô∏è Error en auto-guardado: {e}")

    def _perform_save(self):
        """Realizar guardado de todos los datos registrados"""
        if not self.data_registry:
            return

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_dir = f"backups/auto_save_{timestamp}"
        os.makedirs(backup_dir, exist_ok=True)

        saved_items = 0
        total_size = 0

        try:
            for name, info in self.data_registry.items():
                if info['data'] is not None:
                    filepath = os.path.join(backup_dir, f"{name}.pkl")
                    with open(filepath, 'wb') as f:
                        pickle.dump(info['data'], f)
                    saved_items += 1
                    total_size += info['size_mb']

            metadata = {
                'timestamp': timestamp,
                'items_count': saved_items,
                'total_size_mb': total_size,
                'items_list': list(self.data_registry.keys())
            }

            with open(os.path.join(backup_dir, 'info.json'), 'w') as f:
                json.dump(metadata, f, indent=2)

            self.save_count += 1
            self.last_save = time.time()
            print(f"üíæ Backup #{self.save_count} - {saved_items} elementos ({total_size:.1f}MB)")

        except Exception as e:
            print(f"‚ùå Error guardando backup: {e}")

    def force_save(self):
        """Forzar guardado inmediato"""
        print("üíæ Ejecutando guardado forzado...")
        self._perform_save()
        print("‚úÖ Guardado forzado completado")

# Inicializar sistema de protecci√≥n
os.makedirs('backups', exist_ok=True)
protection_system = SmartAutoSave(autosave_minutes=5)
protection_system.start()

# ============================================================================
# PASO 2: CARGA Y PREPROCESAMIENTO ESPEC√çFICO PARA C√ÅNCER DE PULM√ìN
# ============================================================================

def load_lung_cancer_dataset():
    """Cargar dataset espec√≠fico para c√°ncer de pulm√≥n"""

    print("\nü´Å CARGANDO DATASET DE C√ÅNCER DE PULM√ìN")
    print("=" * 60)

    # Buscar archivo que contenga "CAPULMON" (c√°ncer de pulm√≥n)
    possible_files = []

    # Buscar en diferentes ubicaciones
    search_paths = ['.', '/content', '/content/drive/My Drive', '/content/drive/MyDrive']

    for path in search_paths:
        if os.path.exists(path):
            for file in os.listdir(path):
                if 'CAPULMON' in file.upper() and file.endswith('.csv'):
                    possible_files.append(os.path.join(path, file))

    print(f"üîç Archivos encontrados: {possible_files}")

    if not possible_files:
        print("‚ùå No se encontr√≥ archivo de c√°ncer de pulm√≥n")
        print("üìã Se esperaba archivo con 'CAPULMON' en el nombre")
        return None

    # Intentar cargar el primer archivo encontrado
    df = None
    for file_path in possible_files:
        try:
            print(f"üîç Intentando cargar: {file_path}")

            # Detectar separador y encoding
            separators = ['\t', ',', ';', '|']
            encodings = ['utf-8', 'latin-1', 'cp1252']

            for sep in separators:
                for enc in encodings:
                    try:
                        df_test = pd.read_csv(file_path, delimiter=sep, encoding=enc, nrows=5)
                        if df_test.shape[1] > 1:
                            print(f"‚úÖ Formato detectado: separador='{sep}', encoding='{enc}'")
                            df = pd.read_csv(file_path, delimiter=sep, encoding=enc, low_memory=False)
                            break
                    except:
                        continue
                if df is not None:
                    break

            if df is not None:
                print(f"‚úÖ Dataset cargado: {df.shape[0]:,} filas √ó {df.shape[1]} columnas")
                break

        except Exception as e:
            print(f"‚ùå Error cargando {file_path}: {e}")
            continue

    return df

def preprocess_lung_cancer_data(df):
    """Preprocesamiento espec√≠fico para datos de c√°ncer de pulm√≥n"""

    print("\nüßπ PREPROCESAMIENTO ESPEC√çFICO PARA C√ÅNCER DE PULM√ìN")
    print("-" * 60)

    if df is None:
        return None

    print(f"üìä Dataset original: {df.shape[0]:,} filas √ó {df.shape[1]} columnas")

    # Crear copia
    df_clean = df.copy()

    # Limpiar nombres de columnas
    df_clean.columns = df_clean.columns.str.strip().str.replace(' ', '_')
    print(f"üîß Columnas disponibles: {list(df_clean.columns)}")

    # Convertir tipos de datos
    numeric_columns = ['ENT_RESID', 'EDAD', 'ANIO_REGIS', 'POBLACION']
    for col in numeric_columns:
        if col in df_clean.columns:
            try:
                df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
                print(f"‚úÖ {col} ‚Üí num√©rico")
            except Exception as e:
                print(f"‚ö†Ô∏è Error convirtiendo {col}: {e}")

    # Limpiar datos espec√≠ficos para c√°ncer de pulm√≥n
    initial_rows = len(df_clean)

    # Filtrar edades t√≠picas para c√°ncer de pulm√≥n (m√°s com√∫n en edades avanzadas)
    if 'EDAD' in df_clean.columns:
        # C√°ncer de pulm√≥n t√≠picamente afecta m√°s a partir de los 50 a√±os
        valid_ages = (df_clean['EDAD'] >= 30) & (df_clean['EDAD'] <= 100)
        invalid_ages = (~valid_ages).sum()
        if invalid_ages > 0:
            print(f"üóëÔ∏è Eliminando {invalid_ages:,} registros con edad inv√°lida para c√°ncer pulmonar")
            df_clean = df_clean[valid_ages]

    # Filtrar a√±os v√°lidos
    if 'ANIO_REGIS' in df_clean.columns:
        valid_years = (df_clean['ANIO_REGIS'] >= 2012) & (df_clean['ANIO_REGIS'] <= 2023)
        invalid_years = (~valid_years).sum()
        if invalid_years > 0:
            print(f"üóëÔ∏è Eliminando {invalid_years:,} registros con a√±o inv√°lido")
            df_clean = df_clean[valid_years]

    # Crear variables derivadas espec√≠ficas para c√°ncer de pulm√≥n
    print(f"\nüÜï Creando variables espec√≠ficas para c√°ncer de pulm√≥n...")

    if 'EDAD' in df_clean.columns:
        # Grupos etarios espec√≠ficos para c√°ncer de pulm√≥n
        df_clean['GRUPO_EDAD_PULMON'] = pd.cut(
            df_clean['EDAD'],
            bins=[0, 50, 60, 70, 80, 120],
            labels=['<50', '50-59', '60-69', '70-79', '80+'],
            include_lowest=True
        )
        print("‚úÖ GRUPO_EDAD_PULMON creado (rangos espec√≠ficos)")

        # Categor√≠a de riesgo por edad (c√°ncer de pulm√≥n m√°s com√∫n en edades avanzadas)
        df_clean['RIESGO_EDAD_PULMON'] = df_clean['EDAD'].apply(
            lambda x: 'Alto' if x >= 60 else 'Medio' if x >= 50 else 'Bajo'
        )
        print("‚úÖ RIESGO_EDAD_PULMON creado")

    # An√°lisis espec√≠fico de distribuci√≥n por sexo para c√°ncer de pulm√≥n
    if 'SEXO' in df_clean.columns:
        sexo_dist = df_clean['SEXO'].value_counts()
        print(f"\nüë• DISTRIBUCI√ìN POR SEXO (C√ÅNCER DE PULM√ìN):")
        for sexo, count in sexo_dist.items():
            pct = count / len(df_clean) * 100
            print(f"   {sexo}: {count:,} casos ({pct:.1f}%)")

        # Identificar si hay predominio masculino (t√≠pico en c√°ncer de pulm√≥n)
        if len(sexo_dist) >= 2:
            ratio_hm = sexo_dist.iloc[0] / sexo_dist.iloc[1]
            print(f"   Ratio predominante: {ratio_hm:.2f}:1")

    # Tasa de mortalidad espec√≠fica
    if 'POBLACION' in df_clean.columns and df_clean['POBLACION'].sum() > 0:
        df_clean['TASA_MORTALIDAD_PULMON'] = (1 / df_clean['POBLACION'] * 100000).fillna(0)
        print("‚úÖ TASA_MORTALIDAD_PULMON creada")

    # Per√≠odo temporal
    if 'ANIO_REGIS' in df_clean.columns:
        df_clean['PERIODO_PULMON'] = pd.cut(
            df_clean['ANIO_REGIS'],
            bins=[2011, 2016, 2020, 2023],
            labels=['2012-2016', '2017-2020', '2021-2023'],
            include_lowest=True
        )
        print("‚úÖ PERIODO_PULMON creado")

    # Resumen de limpieza
    final_rows = len(df_clean)
    removed_rows = initial_rows - final_rows

    print(f"\nüìä RESUMEN DE LIMPIEZA ESPEC√çFICA:")
    print(f"   üì• Filas originales: {initial_rows:,}")
    print(f"   üì§ Filas finales: {final_rows:,}")
    print(f"   üóëÔ∏è Filas eliminadas: {removed_rows:,} ({removed_rows/initial_rows*100:.1f}%)")

    return df_clean

# ============================================================================
# PASO 3: AN√ÅLISIS EXPLORATORIO ESPEC√çFICO PARA C√ÅNCER DE PULM√ìN
# ============================================================================

def lung_cancer_exploratory_analysis(df):
    """An√°lisis exploratorio espec√≠fico para c√°ncer de pulm√≥n"""

    print(f"\nüîç AN√ÅLISIS EXPLORATORIO - C√ÅNCER DE PULM√ìN")
    print("=" * 60)

    # Caracter√≠sticas espec√≠ficas del c√°ncer de pulm√≥n
    print(f"üìä CARACTER√çSTICAS EPIDEMIOL√ìGICAS:")
    print(f"   üìè Total de casos: {len(df):,}")

    if 'ANIO_REGIS' in df.columns:
        a√±os = sorted(df['ANIO_REGIS'].unique())
        print(f"   üìÖ Per√≠odo: {min(a√±os)}-{max(a√±os)} ({len(a√±os)} a√±os)")

    if 'ENTIDAD' in df.columns:
        entidades = df['ENTIDAD'].nunique()
        print(f"   üó∫Ô∏è Estados: {entidades} entidades federativas")

    # An√°lisis demogr√°fico espec√≠fico
    if 'EDAD' in df.columns:
        edad_stats = df['EDAD'].describe()
        print(f"\nüë• PERFIL DEMOGR√ÅFICO C√ÅNCER DE PULM√ìN:")
        print(f"   üéØ Edad promedio: {edad_stats['mean']:.1f} a√±os")
        print(f"   üìä Mediana: {edad_stats['50%']:.1f} a√±os")
        print(f"   üìà Rango: {edad_stats['min']:.0f}-{edad_stats['max']:.0f} a√±os")

        # Comparaci√≥n con perfil t√≠pico de c√°ncer de pulm√≥n
        casos_60_mas = (df['EDAD'] >= 60).sum()
        pct_60_mas = casos_60_mas / len(df) * 100
        print(f"   üéØ Casos ‚â•60 a√±os: {casos_60_mas:,} ({pct_60_mas:.1f}%)")
        print(f"   üìù Nota: C√°ncer de pulm√≥n t√≠pico en edades avanzadas")

    # Distribuci√≥n por sexo con interpretaci√≥n
    if 'SEXO' in df.columns:
        sexo_counts = df['SEXO'].value_counts()
        print(f"\nüë• DISTRIBUCI√ìN POR SEXO:")

        total_casos = len(df)
        for sexo, count in sexo_counts.items():
            pct = count / total_casos * 100
            print(f"   {sexo}: {count:,} casos ({pct:.1f}%)")

        # Interpretaci√≥n epidemiol√≥gica
        if len(sexo_counts) >= 2:
            sexo_predominante = sexo_counts.index[0]
            ratio = sexo_counts.iloc[0] / sexo_counts.iloc[1]
            print(f"   üìä Predominio: {sexo_predominante} ({ratio:.2f}:1)")
            print(f"   üìù Nota: Tradicionalmente mayor incidencia masculina en c√°ncer pulmonar")

    # Top estados con mayor incidencia
    if 'ENTIDAD' in df.columns:
        print(f"\nüèÜ TOP 10 ESTADOS - C√ÅNCER DE PULM√ìN:")
        top_estados = df['ENTIDAD'].value_counts().head(10)
        for i, (estado, casos) in enumerate(top_estados.items(), 1):
            pct = casos / len(df) * 100
            print(f"   {i:2d}. {estado}: {casos:,} casos ({pct:.1f}%)")

    # Evoluci√≥n temporal
    if 'ANIO_REGIS' in df.columns:
        casos_anuales = df['ANIO_REGIS'].value_counts().sort_index()
        print(f"\nüìà EVOLUCI√ìN TEMPORAL:")

        if len(casos_anuales) >= 2:
            primer_a√±o = casos_anuales.index[0]
            ultimo_a√±o = casos_anuales.index[-1]
            primer_casos = casos_anuales.iloc[0]
            ultimo_casos = casos_anuales.iloc[-1]

            cambio_absoluto = ultimo_casos - primer_casos
            cambio_pct = (cambio_absoluto / primer_casos) * 100

            print(f"   üìä {primer_a√±o}: {primer_casos:,} casos")
            print(f"   üìä {ultimo_a√±o}: {ultimo_casos:,} casos")
            print(f"   üìà Cambio: {cambio_absoluto:+,} casos ({cambio_pct:+.1f}%)")

    return True

# ============================================================================
# PASO 4: VISUALIZACIONES ESPEC√çFICAS PARA C√ÅNCER DE PULM√ìN
# ============================================================================

def create_lung_cancer_visualizations(df):
    """Crear visualizaciones espec√≠ficas para c√°ncer de pulm√≥n"""

    print(f"\nüìä CREANDO VISUALIZACIONES - C√ÅNCER DE PULM√ìN")
    print("-" * 60)

    # Configurar estilo
    plt.style.use('default')
    sns.set_palette("viridis")

    # Crear figura principal
    fig = plt.figure(figsize=(20, 16))
    fig.suptitle('ü´Å AN√ÅLISIS EPIDEMIOL√ìGICO - C√ÅNCER DE PULM√ìN M√âXICO\nSistema de Machine Learning con Validaci√≥n Temporal',
                 fontsize=18, fontweight='bold', y=0.95)

    plots_created = 0

    # 1. Evoluci√≥n temporal con √©nfasis en tendencias
    if 'ANIO_REGIS' in df.columns:
        try:
            ax1 = plt.subplot(3, 3, 1)
            casos_anuales = df['ANIO_REGIS'].value_counts().sort_index()

            # Gr√°fico principal
            ax1.plot(casos_anuales.index, casos_anuales.values, 'o-',
                    linewidth=3, markersize=8, color='darkred', label='Casos anuales')

            # L√≠nea de tendencia
            if len(casos_anuales) > 2:
                z = np.polyfit(casos_anuales.index, casos_anuales.values, 1)
                p = np.poly1d(z)
                ax1.plot(casos_anuales.index, p(casos_anuales.index),
                        "--", alpha=0.7, color='red', linewidth=2, label='Tendencia')

            ax1.set_title('üìà Evoluci√≥n Temporal - C√°ncer de Pulm√≥n', fontweight='bold')
            ax1.set_xlabel('A√±o')
            ax1.set_ylabel('N√∫mero de Casos')
            ax1.legend()
            ax1.grid(True, alpha=0.3)
            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en gr√°fico temporal: {e}")

    # 2. Distribuci√≥n por edad con rangos espec√≠ficos para c√°ncer pulmonar
    if 'EDAD' in df.columns:
        try:
            ax2 = plt.subplot(3, 3, 2)
            edades = df['EDAD'].dropna()

            # Histograma con √©nfasis en edades t√≠picas de c√°ncer pulmonar
            ax2.hist(edades, bins=30, alpha=0.7, color='lightcoral', edgecolor='black')
            ax2.axvline(edades.mean(), color='red', linestyle='--', linewidth=2,
                       label=f'Media: {edades.mean():.1f} a√±os')
            ax2.axvline(60, color='orange', linestyle=':', linewidth=2,
                       label='Edad t√≠pica de riesgo (60+)')

            ax2.set_title('üë• Distribuci√≥n por Edad - C√°ncer Pulmonar', fontweight='bold')
            ax2.set_xlabel('Edad (a√±os)')
            ax2.set_ylabel('Frecuencia')
            ax2.legend()
            ax2.grid(True, alpha=0.3)
            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en gr√°fico de edad: {e}")

    # 3. Distribuci√≥n por sexo con interpretaci√≥n
    if 'SEXO' in df.columns:
        try:
            ax3 = plt.subplot(3, 3, 3)
            sexo_counts = df['SEXO'].value_counts()

            colors = ['lightblue', 'lightpink'] if len(sexo_counts) == 2 else plt.cm.Set3(range(len(sexo_counts)))
            wedges, texts, autotexts = ax3.pie(sexo_counts.values,
                                              labels=sexo_counts.index,
                                              autopct='%1.1f%%',
                                              colors=colors,
                                              startangle=90)

            ax3.set_title('üë• Distribuci√≥n por Sexo\n(C√°ncer de Pulm√≥n)', fontweight='bold')

            # Agregar nota interpretativa
            if len(sexo_counts) >= 2:
                ratio = sexo_counts.iloc[0] / sexo_counts.iloc[1]
                ax3.text(0, -1.3, f'Ratio: {ratio:.2f}:1', ha='center', fontsize=10,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.5))

            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en gr√°fico de sexo: {e}")

    # 4. Top 10 estados con barras horizontales
    if 'ENTIDAD' in df.columns:
        try:
            ax4 = plt.subplot(3, 3, 4)
            top_estados = df['ENTIDAD'].value_counts().head(10)

            bars = ax4.barh(range(len(top_estados)), top_estados.values, color='darkgreen')
            ax4.set_yticks(range(len(top_estados)))
            ax4.set_yticklabels([str(x)[:15] + '...' if len(str(x)) > 15 else str(x)
                                for x in top_estados.index])
            ax4.set_title('üó∫Ô∏è Top 10 Estados - C√°ncer Pulm√≥n', fontweight='bold')
            ax4.set_xlabel('N√∫mero de Casos')
            ax4.grid(True, alpha=0.3, axis='x')

            # Agregar valores en las barras
            for i, bar in enumerate(bars):
                width = bar.get_width()
                ax4.text(width + width*0.01, bar.get_y() + bar.get_height()/2,
                        f'{int(width):,}', ha='left', va='center', fontsize=8)

            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en gr√°fico de estados: {e}")

    # 5. Grupos etarios espec√≠ficos para c√°ncer pulmonar
    if 'GRUPO_EDAD_PULMON' in df.columns:
        try:
            ax5 = plt.subplot(3, 3, 5)
            grupo_counts = df['GRUPO_EDAD_PULMON'].value_counts().sort_index()

            bars = ax5.bar(range(len(grupo_counts)), grupo_counts.values,
                          color='coral', alpha=0.8, edgecolor='black')
            ax5.set_xticks(range(len(grupo_counts)))
            ax5.set_xticklabels(grupo_counts.index, rotation=45)
            ax5.set_title('üìä Grupos Etarios Espec√≠ficos\n(C√°ncer Pulmonar)', fontweight='bold')
            ax5.set_ylabel('N√∫mero de Casos')
            ax5.grid(True, alpha=0.3, axis='y')

            # Resaltar grupo de mayor riesgo (60+)
            for i, bar in enumerate(bars):
                if '60' in str(grupo_counts.index[i]) or '70' in str(grupo_counts.index[i]) or '80' in str(grupo_counts.index[i]):
                    bar.set_color('red')
                    bar.set_alpha(0.9)

            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en grupos etarios: {e}")

    # 6. Evoluci√≥n por per√≠odos
    if 'PERIODO_PULMON' in df.columns:
        try:
            ax6 = plt.subplot(3, 3, 6)
            periodo_counts = df['PERIODO_PULMON'].value_counts().sort_index()

            wedges, texts, autotexts = ax6.pie(periodo_counts.values,
                                              labels=periodo_counts.index,
                                              autopct='%1.1f%%',
                                              colors=['lightblue', 'lightgreen', 'lightcoral'],
                                              startangle=90)
            ax6.set_title('üìä Distribuci√≥n por Per√≠odos\n(C√°ncer Pulmonar)', fontweight='bold')
            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en per√≠odos: {e}")

    # 7. Comparaci√≥n de incidencia por sexo a lo largo del tiempo
    if 'ANIO_REGIS' in df.columns and 'SEXO' in df.columns:
        try:
            ax7 = plt.subplot(3, 3, 7)

            sexos_unicos = df['SEXO'].unique()
            colors = ['blue', 'red', 'green']

            for i, sexo in enumerate(sexos_unicos[:3]):
                sexo_data = df[df['SEXO'] == sexo]
                casos_anuales_sexo = sexo_data['ANIO_REGIS'].value_counts().sort_index()

                if len(casos_anuales_sexo) > 1:
                    ax7.plot(casos_anuales_sexo.index, casos_anuales_sexo.values,
                            'o-', label=f'{sexo}', color=colors[i], linewidth=2, markersize=4)

            ax7.set_title('üìä Evoluci√≥n por Sexo\n(C√°ncer Pulmonar)', fontweight='bold')
            ax7.set_xlabel('A√±o')
            ax7.set_ylabel('Casos')
            ax7.legend()
            ax7.grid(True, alpha=0.3)
            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en evoluci√≥n por sexo: {e}")

    # 8. Concentraci√≥n geogr√°fica (an√°lisis de Pareto)
    if 'ENTIDAD' in df.columns:
        try:
            ax8 = plt.subplot(3, 3, 8)
            entidad_counts = df['ENTIDAD'].value_counts()

            # Calcular porcentaje acumulado
            porcentaje_acum = (entidad_counts.cumsum() / entidad_counts.sum() * 100)

            # Mostrar top 10
            bars = ax8.bar(range(min(10, len(entidad_counts))),
                          entidad_counts.head(10).values, alpha=0.7, color='lightblue')

            # L√≠nea de Pareto
            ax8_twin = ax8.twinx()
            ax8_twin.plot(range(min(10, len(porcentaje_acum))),
                         porcentaje_acum.head(10).values, 'ro-', color='red')
            ax8_twin.set_ylabel('% Acumulado', color='red')

            ax8.set_title('üìä Concentraci√≥n Geogr√°fica\n(An√°lisis 80/20)', fontweight='bold')
            ax8.set_xlabel('Top Estados')
            ax8.set_ylabel('Casos')
            ax8.grid(True, alpha=0.3)
            plots_created += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error en an√°lisis de Pareto: {e}")

    # 9. Resumen de m√©tricas clave
    try:
        ax9 = plt.subplot(3, 3, 9)

        # M√©tricas espec√≠ficas para c√°ncer de pulm√≥n
        metricas = []
        valores = []

        metricas.append('Total Casos')
        valores.append(len(df))

        if 'EDAD' in df.columns:
            metricas.append('Edad Media')
            valores.append(round(df['EDAD'].mean(), 1))

            # Casos en edad de alto riesgo (60+)
            casos_alto_riesgo = (df['EDAD'] >= 60).sum()
            metricas.append('Casos 60+ a√±os')
            valores.append(casos_alto_riesgo)

        if 'ENTIDAD' in df.columns:
            metricas.append('Estados')
            valores.append(df['ENTIDAD'].nunique())

        # Gr√°fico de barras horizontal
        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'][:len(metricas)]
        bars = ax9.barh(metricas, valores, color=colors)

        ax9.set_title('üìà M√©tricas Clave\nC√°ncer de Pulm√≥n', fontweight='bold')
        ax9.set_xlabel('Valor')

        # Agregar valores en las barras
        for i, (bar, valor) in enumerate(zip(bars, valores)):
            ax9.text(bar.get_width() + max(valores)*0.01, bar.get_y() + bar.get_height()/2,
                    f'{valor:,}' if isinstance(valor, int) else f'{valor}',
                    ha='left', va='center', fontweight='bold')

        plots_created += 1

    except Exception as e:
        print(f"‚ö†Ô∏è Error en m√©tricas clave: {e}")

    plt.tight_layout()
    plt.show()

    print(f"‚úÖ Visualizaciones creadas: {plots_created}/9")
    return plots_created

# ============================================================================
# PASO 5: MODELOS DE ML ESPEC√çFICOS PARA C√ÅNCER DE PULM√ìN
# ============================================================================

class LungCancerRiskClassification:
    """Modelo de clasificaci√≥n de riesgo espec√≠fico para c√°ncer de pulm√≥n"""

    def __init__(self):
        self.models = {}
        self.encoders = {}
        self.scaler = StandardScaler()
        self.temporal_metrics = {}
        self.results = {}

    def prepare_lung_cancer_features(self, df, period_filter=None):
        """Preparar caracter√≠sticas espec√≠ficas para c√°ncer de pulm√≥n"""

        if period_filter:
            df_filtered = df[df['ANIO_REGIS'].isin(period_filter)]
            print(f"üìä Preparando features para per√≠odo {min(period_filter)}-{max(period_filter)}")
        else:
            df_filtered = df

        # M√©tricas por estado espec√≠ficas para c√°ncer pulmonar
        state_features = df_filtered.groupby('ENTIDAD').agg({
            'EDAD': ['mean', 'std', 'count'],
            'ANIO_REGIS': ['count', 'min', 'max'],
            'POBLACION': ['mean', 'sum']
        }).round(3)

        # Aplanar columnas
        state_features.columns = ['_'.join(col) for col in state_features.columns]
        state_features = state_features.reset_index()

        # Variables espec√≠ficas para c√°ncer de pulm√≥n
        state_features['casos_por_a√±o'] = (
            state_features['ANIO_REGIS_count'] /
            (state_features['ANIO_REGIS_max'] - state_features['ANIO_REGIS_min'] + 1)
        ).fillna(0)

        state_features['densidad_casos_pulmon'] = (
            state_features['ANIO_REGIS_count'] / state_features['POBLACION_sum'] * 100000
        ).fillna(0)

        # Factor de riesgo por edad (c√°ncer pulmonar m√°s com√∫n en edades avanzadas)
        state_features['factor_edad_pulmon'] = (
            (state_features['EDAD_mean'] >= 60).astype(int) * 0.4 +
            (state_features['EDAD_mean'] >= 55).astype(int) * 0.3
        )

        # Score de riesgo espec√≠fico para c√°ncer pulmonar
        state_features['risk_score_pulmon'] = (
            state_features['casos_por_a√±o'] * 0.3 +
            state_features['densidad_casos_pulmon'] * 0.3 +
            state_features['factor_edad_pulmon'] * 0.4
        )

        # Categorizar riesgo en 3 niveles
        state_features['risk_category_pulmon'] = pd.qcut(
            state_features['risk_score_pulmon'],
            q=3,
            labels=['Bajo', 'Medio', 'Alto'],
            duplicates='drop'
        )

        return state_features

    def train_lung_cancer_models(self, df):
        """Entrenar modelos espec√≠ficos para c√°ncer de pulm√≥n con validaci√≥n temporal"""

        print("\nü´Å MODELO 1: CLASIFICACI√ìN DE RIESGO - C√ÅNCER DE PULM√ìN")
        print("-" * 65)

        # Divisi√≥n temporal espec√≠fica para c√°ncer pulmonar
        train_years = list(range(2012, 2021))  # 2012-2020
        test_years = list(range(2021, 2024))   # 2021-2023

        print(f"üìä Entrenamiento: {min(train_years)}-{max(train_years)}")
        print(f"üéØ Validaci√≥n temporal: {min(test_years)}-{max(test_years)}")

        # Preparar features
        train_features = self.prepare_lung_cancer_features(df, train_years)
        test_features = self.prepare_lung_cancer_features(df, test_years)

        # Estados comunes
        common_states = set(train_features['ENTIDAD']) & set(test_features['ENTIDAD'])
        print(f"üó∫Ô∏è Estados con datos en ambos per√≠odos: {len(common_states)}")

        # Filtrar estados comunes
        train_filtered = train_features[train_features['ENTIDAD'].isin(common_states)]
        test_filtered = test_features[test_features['ENTIDAD'].isin(common_states)]

        # Features espec√≠ficas para c√°ncer pulmonar
        feature_cols = ['casos_por_a√±o', 'densidad_casos_pulmon', 'EDAD_mean', 'factor_edad_pulmon']

        X_train = train_filtered[feature_cols].fillna(0)
        y_train = train_filtered['risk_category_pulmon'].fillna('Medio')

        X_test = test_filtered[feature_cols].fillna(0)
        y_test = test_filtered['risk_category_pulmon'].fillna('Medio')

        # Codificar target
        le = LabelEncoder()
        y_train_encoded = le.fit_transform(y_train)
        y_test_encoded = le.transform(y_test)
        self.encoders['risk_target'] = le

        # Escalar features
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)

        # Modelos espec√≠ficos para c√°ncer pulmonar
        models_config = {
            'RandomForest_Pulmon': RandomForestClassifier(
                n_estimators=150,
                max_depth=10,
                min_samples_split=5,
                random_state=42
            ),
            'GradientBoosting_Pulmon': GradientBoostingRegressor(
                n_estimators=100,
                learning_rate=0.1,
                max_depth=6,
                random_state=42
            ),
            'NaiveBayes_Pulmon': GaussianNB()
        }

        # Entrenar modelos
        results = {}

        for name, model in models_config.items():
            try:
                print(f"ü§ñ Entrenando {name}...")

                if 'GradientBoosting' in name:
                    # Para regresores
                    model.fit(X_train_scaled, y_train_encoded)
                    train_pred = model.predict(X_train_scaled)
                    test_pred = model.predict(X_test_scaled)

                    train_pred_class = np.round(np.clip(train_pred, 0, 2)).astype(int)
                    test_pred_class = np.round(np.clip(test_pred, 0, 2)).astype(int)

                    train_score = np.mean(train_pred_class == y_train_encoded)
                    temporal_score = np.mean(test_pred_class == y_test_encoded)
                else:
                    # Para clasificadores
                    model.fit(X_train_scaled, y_train_encoded)
                    train_score = model.score(X_train_scaled, y_train_encoded)
                    temporal_score = model.score(X_test_scaled, y_test_encoded)

                # Cross-validation
                cv_scores = cross_val_score(
                    RandomForestClassifier(random_state=42),
                    X_train_scaled, y_train_encoded, cv=3
                )

                results[name] = {
                    'model': model,
                    'train_score': train_score,
                    'temporal_score': temporal_score,
                    'cv_mean': cv_scores.mean(),
                    'cv_std': cv_scores.std()
                }

                print(f"   üìä Train: {train_score:.3f}")
                print(f"   ‚è∞ TEMPORAL: {temporal_score:.3f}")
                print(f"   üîÑ CV: {cv_scores.mean():.3f}¬±{cv_scores.std():.3f}")

            except Exception as e:
                print(f"   ‚ùå Error entrenando {name}: {e}")
                continue

        # Seleccionar mejor modelo
        if results:
            best_model_name = max(results.keys(), key=lambda k: results[k]['temporal_score'])
            self.models['best_classifier'] = results[best_model_name]['model']

            self.temporal_metrics = {
                'best_model': best_model_name,
                'temporal_accuracy': results[best_model_name]['temporal_score'],
                'train_accuracy': results[best_model_name]['train_score'],
                'cv_score': results[best_model_name]['cv_mean'],
                'states_evaluated': len(common_states),
                'train_period': f"{min(train_years)}-{max(train_years)}",
                'test_period': f"{min(test_years)}-{max(test_years)}",
                'cancer_type': 'C√°ncer de Pulm√≥n'
            }

            print(f"\nüèÜ MEJOR MODELO PARA C√ÅNCER PULMONAR: {best_model_name}")
            print(f"üéØ Precisi√≥n temporal: {results[best_model_name]['temporal_score']:.3f}")
            print(f"üìä Estados evaluados: {len(common_states)}")

            self.results = results
            return results
        else:
            print("‚ùå No se pudo entrenar ning√∫n modelo")
            return None

class LungCancerTimeSeriesModel:
    """Modelo de series temporales espec√≠fico para c√°ncer de pulm√≥n"""

    def __init__(self):
        self.models = {}
        self.predictions = {}
        self.temporal_validation = {}

    def prepare_lung_cancer_timeseries(self, df):
        """Preparar series temporales espec√≠ficas para c√°ncer de pulm√≥n"""

        print("\nüìà MODELO 2: SERIES TEMPORALES - C√ÅNCER DE PULM√ìN")
        print("-" * 60)

        # Datos agregados por a√±o con an√°lisis espec√≠fico
        ts_data = df.groupby('ANIO_REGIS').agg({
            'EDAD': ['count', 'mean'],  # Casos y edad promedio
            'POBLACION': 'sum'
        }).reset_index()

        ts_data.columns = ['a√±o', 'casos', 'edad_promedio', 'poblacion_total']
        ts_data['tasa_casos_pulmon'] = ts_data['casos'] / ts_data['poblacion_total'] * 100000

        # An√°lisis de tendencias espec√≠ficas para c√°ncer pulmonar
        if 'SEXO' in df.columns:
            # Casos por sexo
            casos_sexo = df.groupby(['ANIO_REGIS', 'SEXO']).size().unstack(fill_value=0)
            ts_data = ts_data.merge(
                casos_sexo.reset_index().rename(columns={'ANIO_REGIS': 'a√±o'}),
                on='a√±o', how='left'
            )

        print(f"üìä Serie temporal: {len(ts_data)} a√±os de datos")

        # Divisi√≥n temporal
        train_data = ts_data[ts_data['a√±o'] <= 2020].copy()
        test_data = ts_data[ts_data['a√±o'] >= 2021].copy()

        print(f"üìä Entrenamiento: {len(train_data)} a√±os ({train_data['a√±o'].min()}-{train_data['a√±o'].max()})")
        print(f"üéØ Validaci√≥n: {len(test_data)} a√±os ({test_data['a√±o'].min()}-{test_data['a√±o'].max()})")

        return {
            'complete': ts_data,
            'train': train_data,
            'test': test_data
        }

    def train_lung_cancer_timeseries(self, ts_data_dict):
        """Entrenar modelos de series temporales para c√°ncer de pulm√≥n"""

        train_data = ts_data_dict['train']
        test_data = ts_data_dict['test']

        if len(test_data) == 0:
            print("‚ö†Ô∏è No hay datos de validaci√≥n temporal")
            return None

        print("ü§ñ Entrenando modelos de series temporales para c√°ncer pulmonar...")

        X_train = train_data['a√±o'].values.reshape(-1, 1)
        y_train = train_data['casos'].values

        X_test = test_data['a√±o'].values.reshape(-1, 1)
        y_test = test_data['casos'].values

        # Modelo lineal mejorado para c√°ncer pulmonar
        linear_model = LinearRegression()
        linear_model.fit(X_train, y_train)

        # Predicciones
        train_pred = linear_model.predict(X_train)
        test_pred = linear_model.predict(X_test)

        # M√©tricas
        train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
        test_rmse = np.sqrt(mean_squared_error(y_test, test_pred))
        test_r2 = r2_score(y_test, test_pred)

        print(f"   üìä Train RMSE: {train_rmse:.2f}")
        print(f"   ‚è∞ Test RMSE: {test_rmse:.2f}")
        print(f"   üìà Test R¬≤: {test_r2:.3f}")

        # Guardar modelo
        self.models['linear_pulmon'] = linear_model

        # Proyecciones futuras espec√≠ficas para c√°ncer pulmonar
        future_years = np.array(range(2024, 2031)).reshape(-1, 1)
        future_pred = linear_model.predict(future_years)

        # Ajustar predicciones considerando tendencias del c√°ncer pulmonar
        # (t√≠picamente m√°s estables que otros tipos de c√°ncer)
        future_pred_adjusted = np.maximum(0, future_pred * 0.95)  # Factor conservador

        self.predictions['future_cases_pulmon'] = pd.DataFrame({
            'a√±o': range(2024, 2031),
            'casos_proyectados_pulmon': future_pred_adjusted.astype(int),
            'casos_original': future_pred.astype(int)
        })

        print(f"üîÆ Proyecciones para c√°ncer pulmonar 2024-2030:")
        for _, row in self.predictions['future_cases_pulmon'].iterrows():
            print(f"   {row['a√±o']}: {row['casos_proyectados_pulmon']:,} casos")

        self.temporal_validation = {
            'model_type': 'linear_regression_pulmon',
            'train_rmse': train_rmse,
            'test_rmse': test_rmse,
            'test_r2': test_r2,
            'cancer_type': 'C√°ncer de Pulm√≥n',
            'projection_years': list(range(2024, 2031))
        }

        return self.temporal_validation

class LungCancerGeographicClustering:
    """Clustering geogr√°fico espec√≠fico para c√°ncer de pulm√≥n"""

    def __init__(self):
        self.models = {}
        self.cluster_results = {}

    def perform_lung_cancer_clustering(self, df):
        """Clustering espec√≠fico para patrones de c√°ncer de pulm√≥n"""

        print("\nüó∫Ô∏è MODELO 3: CLUSTERING GEOGR√ÅFICO - C√ÅNCER DE PULM√ìN")
        print("-" * 65)

        # Features espec√≠ficas para c√°ncer pulmonar por estado
        state_features = df.groupby('ENTIDAD').agg({
            'EDAD': ['mean', 'std'],
            'ANIO_REGIS': 'count',
            'POBLACION': 'mean'
        }).round(3)

        state_features.columns = ['edad_media', 'edad_std', 'total_casos', 'poblacion_media']
        state_features = state_features.reset_index()

        # Variables espec√≠ficas para clustering de c√°ncer pulmonar
        state_features['casos_per_capita_pulmon'] = (
            state_features['total_casos'] / state_features['poblacion_media'] * 100000
        )

        # Factor de riesgo por edad (importante en c√°ncer pulmonar)
        state_features['factor_edad_alto_riesgo'] = (
            (state_features['edad_media'] >= 60).astype(int)
        )

        # √çndice de concentraci√≥n de casos
        state_features['indice_concentracion'] = (
            state_features['total_casos'] / state_features['total_casos'].sum()
        )

        print(f"üó∫Ô∏è Estados a analizar: {len(state_features)}")

        # Features para clustering
        feature_cols = [
            'edad_media', 'total_casos', 'casos_per_capita_pulmon',
            'factor_edad_alto_riesgo', 'indice_concentracion'
        ]
        X = state_features[feature_cols].fillna(state_features[feature_cols].mean())

        # Normalizar
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Determinar n√∫mero √≥ptimo de clusters
        silhouette_scores = {}
        for n_clusters in range(2, 6):
            kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
            cluster_labels = kmeans.fit_predict(X_scaled)
            silhouette_avg = silhouette_score(X_scaled, cluster_labels)
            silhouette_scores[n_clusters] = silhouette_avg
            print(f"üìä {n_clusters} clusters - Silhouette: {silhouette_avg:.3f}")

        # Entrenar modelo final
        best_n_clusters = max(silhouette_scores.keys(), key=lambda k: silhouette_scores[k])
        final_kmeans = KMeans(n_clusters=best_n_clusters, random_state=42, n_init=10)
        cluster_labels = final_kmeans.fit_predict(X_scaled)

        state_features['cluster_pulmon'] = cluster_labels

        # An√°lisis de clusters espec√≠fico para c√°ncer pulmonar
        cluster_analysis = {}
        cluster_names = ['Riesgo Bajo', 'Riesgo Medio', 'Riesgo Alto', 'Riesgo Muy Alto']

        for cluster_id in range(best_n_clusters):
            cluster_states = state_features[state_features['cluster_pulmon'] == cluster_id]

            cluster_analysis[cluster_id] = {
                'nombre': cluster_names[cluster_id] if cluster_id < len(cluster_names) else f'Cluster {cluster_id}',
                'estados': cluster_states['ENTIDAD'].tolist(),
                'num_estados': len(cluster_states),
                'casos_promedio': cluster_states['total_casos'].mean(),
                'edad_promedio': cluster_states['edad_media'].mean(),
                'casos_per_capita_promedio': cluster_states['casos_per_capita_pulmon'].mean(),
                'factor_alto_riesgo': cluster_states['factor_edad_alto_riesgo'].mean()
            }

            print(f"\nüìä {cluster_analysis[cluster_id]['nombre']}:")
            print(f"   Estados: {len(cluster_states)}")
            print(f"   Casos promedio: {cluster_states['total_casos'].mean():.1f}")
            print(f"   Edad promedio: {cluster_states['edad_media'].mean():.1f} a√±os")
            print(f"   Casos per c√°pita: {cluster_states['casos_per_capita_pulmon'].mean():.2f}")

        self.models['kmeans_pulmon'] = final_kmeans
        self.models['scaler_pulmon'] = scaler
        self.cluster_results = {
            'state_features': state_features,
            'cluster_analysis': cluster_analysis,
            'best_n_clusters': best_n_clusters,
            'silhouette_scores': silhouette_scores,
            'cancer_type': 'C√°ncer de Pulm√≥n'
        }

        print(f"\n‚úÖ Clustering de c√°ncer pulmonar completado con {best_n_clusters} clusters")

        return cluster_analysis

# ============================================================================
# PASO 6: EJECUTAR AN√ÅLISIS COMPLETO PARA C√ÅNCER DE PULM√ìN
# ============================================================================

def run_complete_lung_cancer_analysis():
    """Ejecutar an√°lisis completo espec√≠fico para c√°ncer de pulm√≥n"""

    print("ü´Å EJECUTANDO AN√ÅLISIS COMPLETO - C√ÅNCER DE PULM√ìN")
    print("=" * 70)

    # 1. Cargar datos
    df_raw = load_lung_cancer_dataset()
    if df_raw is None:
        print("‚ùå No se pudo cargar el dataset de c√°ncer de pulm√≥n")
        return None

    # Proteger datos originales
    protection_system.register_data('dataset_pulmon_raw', df_raw)

    # 2. Preprocesar datos
    df_clean = preprocess_lung_cancer_data(df_raw)
    if df_clean is None:
        print("‚ùå Error en preprocesamiento")
        return None

    # Proteger datos limpios
    protection_system.register_data('dataset_pulmon_clean', df_clean)

    # 3. An√°lisis exploratorio
    lung_cancer_exploratory_analysis(df_clean)

    # 4. Crear visualizaciones
    create_lung_cancer_visualizations(df_clean)

    # 5. Modelos de Machine Learning
    results_ml = {}

    # Modelo 1: Clasificaci√≥n de riesgo
    risk_model = LungCancerRiskClassification()
    risk_results = risk_model.train_lung_cancer_models(df_clean)
    if risk_results:
        results_ml['risk_classification_pulmon'] = {
            'model': risk_model,
            'results': risk_results,
            'temporal_metrics': risk_model.temporal_metrics
        }

    # Modelo 2: Series temporales
    ts_model = LungCancerTimeSeriesModel()
    ts_data = ts_model.prepare_lung_cancer_timeseries(df_clean)
    ts_validation = ts_model.train_lung_cancer_timeseries(ts_data)
    if ts_validation:
        results_ml['time_series_pulmon'] = {
            'model': ts_model,
            'validation': ts_validation,
            'predictions': ts_model.predictions
        }

    # Modelo 3: Clustering geogr√°fico
    cluster_model = LungCancerGeographicClustering()
    cluster_analysis = cluster_model.perform_lung_cancer_clustering(df_clean)
    if cluster_analysis:
        results_ml['clustering_pulmon'] = {
            'model': cluster_model,
            'analysis': cluster_analysis
        }

    # Proteger resultados
    protection_system.register_data('ml_results_pulmon', results_ml)
    protection_system.register_data('dataset_final_pulmon', df_clean)
    protection_system.force_save()

    print(f"\nüõ°Ô∏è Todos los datos protegidos autom√°ticamente")

    return {
        'dataset_raw': df_raw,
        'dataset_clean': df_clean,
        'ml_results': results_ml,
        'cancer_type': 'C√°ncer de Pulm√≥n'
    }

# ============================================================================
# PASO 7: REPORTE FINAL ESPEC√çFICO PARA C√ÅNCER DE PULM√ìN
# ============================================================================

def generate_lung_cancer_final_report(analysis_results):
    """Generar reporte final espec√≠fico para c√°ncer de pulm√≥n"""

    print(f"\nüìã REPORTE FINAL - C√ÅNCER DE PULM√ìN M√âXICO")
    print("=" * 70)

    if not analysis_results:
        print("‚ùå No hay resultados para reportar")
        return

    df = analysis_results['dataset_clean']
    ml_results = analysis_results['ml_results']

    # Encabezado espec√≠fico
    fecha_reporte = datetime.now().strftime("%d de %B de %Y, %H:%M")
    print(f"ü´Å SISTEMA DE MACHINE LEARNING EPIDEMIOL√ìGICO")
    print(f"üìä AN√ÅLISIS DE C√ÅNCER DE PULM√ìN EN M√âXICO")
    print(f"üìÖ Fecha del reporte: {fecha_reporte}")

    # 1. RESUMEN EJECUTIVO ESPEC√çFICO
    print(f"\nüéØ RESUMEN EJECUTIVO - C√ÅNCER DE PULM√ìN")
    print("-" * 50)

    total_registros = len(df)
    print(f"üìä Casos analizados: {total_registros:,} registros de c√°ncer pulmonar")

    if 'ANIO_REGIS' in df.columns:
        a√±o_min = df['ANIO_REGIS'].min()
        a√±o_max = df['ANIO_REGIS'].max()
        print(f"üìÖ Per√≠odo de an√°lisis: {a√±o_min}-{a√±o_max}")

    if 'ENTIDAD' in df.columns:
        num_entidades = df['ENTIDAD'].nunique()
        print(f"üó∫Ô∏è Cobertura: {num_entidades} entidades federativas")

    # Caracter√≠sticas espec√≠ficas del c√°ncer pulmonar
    if 'EDAD' in df.columns:
        edad_promedio = df['EDAD'].mean()
        casos_60_mas = (df['EDAD'] >= 60).sum()
        pct_60_mas = casos_60_mas / len(df) * 100
        print(f"üë• Edad promedio: {edad_promedio:.1f} a√±os")
        print(f"üéØ Casos ‚â•60 a√±os: {casos_60_mas:,} ({pct_60_mas:.1f}%)")

    # Distribuci√≥n por sexo
    if 'SEXO' in df.columns:
        sexo_dist = df['SEXO'].value_counts()
        print(f"üë• Distribuci√≥n por sexo:")
        for sexo, count in sexo_dist.items():
            pct = count / len(df) * 100
            print(f"   {sexo}: {count:,} ({pct:.1f}%)")

    # 2. HALLAZGOS ESPEC√çFICOS DE C√ÅNCER PULMONAR
    print(f"\nüîç HALLAZGOS ESPEC√çFICOS - C√ÅNCER DE PULM√ìN")
    print("-" * 55)

    print(f"üìà Caracter√≠sticas epidemiol√≥gicas:")
    print(f"   ‚Ä¢ Mayor incidencia en edades avanzadas (‚â•60 a√±os)")
    print(f"   ‚Ä¢ Patr√≥n de distribuci√≥n geogr√°fica identificado")

    if ml_results:
        print(f"ü§ñ Resultados de Machine Learning:")

        if 'risk_classification_pulmon' in ml_results:
            risk_metrics = ml_results['risk_classification_pulmon']['temporal_metrics']
            print(f"   üéØ Clasificaci√≥n de Riesgo Estados: {risk_metrics['temporal_accuracy']:.3f}")

        if 'time_series_pulmon' in ml_results:
            ts_validation = ml_results['time_series_pulmon']['validation']
            print(f"   üìà Proyecciones Temporales R¬≤: {ts_validation['test_r2']:.3f}")

        if 'clustering_pulmon' in ml_results:
            cluster_results = ml_results['clustering_pulmon']['model'].cluster_results
            print(f"   üó∫Ô∏è Clusters Geogr√°ficos: {cluster_results['best_n_clusters']} grupos identificados")

    # 3. PROYECCIONES Y RECOMENDACIONES ESPEC√çFICAS
    print(f"\nüîÆ PROYECCIONES ESPEC√çFICAS - C√ÅNCER DE PULM√ìN")
    print("-" * 55)

    if ml_results and 'time_series_pulmon' in ml_results:
        predictions = ml_results['time_series_pulmon']['predictions']
        if 'future_cases_pulmon' in predictions:
            future_data = predictions['future_cases_pulmon']
            print(f"üìä Proyecciones 2024-2030:")
            for _, row in future_data.head(3).iterrows():
                print(f"   {row['a√±o']}: {row['casos_proyectados_pulmon']:,} casos estimados")
            print(f"   ... (proyecciones completas hasta 2030)")

    # 4. RECOMENDACIONES ESPEC√çFICAS PARA C√ÅNCER PULMONAR
    print(f"\nüí° RECOMENDACIONES ESPEC√çFICAS - C√ÅNCER DE PULM√ìN")
    print("-" * 60)

    recomendaciones_pulmon = [
        "üö≠ Intensificar programas de cesaci√≥n tab√°quica en poblaciones de riesgo",
        "üè≠ Vigilancia epidemiol√≥gica en zonas industriales y urbanas",
        "üë¥ Programas de detecci√≥n temprana en poblaci√≥n ‚â•50 a√±os",
        "ü´Å Implementar tomograf√≠a de baja dosis en grupos de alto riesgo",
        "üìä Monitoreo espec√≠fico de tendencias por grupos etarios",
        "üó∫Ô∏è Focalizar recursos en estados de alta incidencia identificados",
        "ü§ñ Integrar modelos predictivos en sistemas de vigilancia epidemiol√≥gica",
        "üë• Estrategias diferenciadas considerando perfil predominantemente masculino"
    ]

    for i, rec in enumerate(recomendaciones_pulmon, 1):
        print(f"{i}. {rec}")

    # 5. COMPARACI√ìN CON C√ÅNCER DE MAMA
    print(f"\nüî¨ DIFERENCIAS EPIDEMIOL√ìGICAS VS C√ÅNCER DE MAMA")
    print("-" * 60)

    diferencias = [
        "üë• Distribuci√≥n por sexo: Mayor incidencia masculina vs femenina",
        "üìÖ Edad de inicio: T√≠picamente m√°s tard√≠a (60+ vs 50+)",
        "üéØ Factores de riesgo: Tabaquismo/exposici√≥n vs hormonales",
        "üìà Tendencias: Patrones de incidencia diferentes",
        "üó∫Ô∏è Distribuci√≥n geogr√°fica: Posible correlaci√≥n con zonas industriales",
        "üîÆ Pron√≥stico: Diferente evoluci√≥n temporal esperada"
    ]

    for diferencia in diferencias:
        print(f"   ‚Ä¢ {diferencia}")

    print(f"\n" + "üéâ" + "="*68 + "üéâ")
    print("     ¬°AN√ÅLISIS DE C√ÅNCER DE PULM√ìN COMPLETADO EXITOSAMENTE!")
    print("üéâ" + "="*68 + "üéâ")

    print(f"""
ü´Å SISTEMA ML ESPEC√çFICO PARA C√ÅNCER DE PULM√ìN implementado con:

‚úÖ An√°lisis de {total_registros:,} registros reales
‚úÖ Modelos adaptados a caracter√≠sticas epidemiol√≥gicas espec√≠ficas
‚úÖ Validaci√≥n temporal robusta (2012-2020 vs 2021-2023)
‚úÖ Consideraci√≥n de factores de riesgo espec√≠ficos (edad, sexo)
‚úÖ Clustering geogr√°fico adaptado a patrones pulmonares
‚úÖ Proyecciones espec√≠ficas 2024-2030
‚úÖ Recomendaciones targeted para c√°ncer de pulm√≥n

üéØ LISTO PARA: Vigilancia epidemiol√≥gica especializada,
              pol√≠ticas de salud p√∫blica espec√≠ficas,
              y toma de decisiones en c√°ncer pulmonar.
""")

# ============================================================================
# EJECUCI√ìN PRINCIPAL
# ============================================================================

def main():
    """Funci√≥n principal para ejecutar an√°lisis completo de c√°ncer de pulm√≥n"""

    print("üöÄ INICIANDO SISTEMA ML - C√ÅNCER DE PULM√ìN M√âXICO")
    print("=" * 70)
    print("üìä Versi√≥n adaptada espec√≠ficamente para c√°ncer pulmonar")
    print("üîÑ Considerando diferencias epidemiol√≥gicas clave")

    try:
        # Ejecutar an√°lisis completo
        results = run_complete_lung_cancer_analysis()

        if results:
            # Generar reporte final
            generate_lung_cancer_final_report(results)

            # Mostrar comandos √∫tiles
            print(f"\nüíª COMANDOS √öTILES:")
            print("=" * 30)
            print("protection_system.get_detailed_status()  # Ver estado")
            print("protection_system.manual_save()         # Guardar manual")
            print("results['dataset_clean'].head()         # Ver datos")
            print("results['ml_results']                    # Ver resultados ML")

            print(f"\nüéØ VARIABLES DISPONIBLES:")
            print("   results['dataset_clean']           ‚Üí Dataset procesado")
            print("   results['ml_results']              ‚Üí Modelos entrenados")
            print("   results['cancer_type']             ‚Üí Tipo de c√°ncer")

            return results

        else:
            print("‚ùå Error en el an√°lisis")
            return None

    except Exception as e:
        print(f"‚ùå Error general: {e}")
        print("üõ°Ô∏è Intentando guardar trabajo...")
        try:
            protection_system.force_save()
            print("‚úÖ Trabajo guardado en backup")
        except:
            print("‚ö†Ô∏è No se pudo guardar backup")
        return None

# ============================================================================
# INSTRUCCIONES DE USO
# ============================================================================

print("üìã INSTRUCCIONES DE USO:")
print("=" * 30)
print("1. Aseg√∫rate de tener el archivo DATA_CAPULMON_2012a2023.csv")
print("2. Ejecuta: results = main()")
print("3. El sistema se adaptar√° autom√°ticamente a c√°ncer de pulm√≥n")
print("4. Revisa las diferencias vs el an√°lisis de c√°ncer de mama")
print("5. Todos los datos se protegen autom√°ticamente")

print(f"\nüîÑ DIFERENCIAS CLAVE IMPLEMENTADAS:")
print("   üéØ Rangos etarios espec√≠ficos para c√°ncer pulmonar")
print("   üë• An√°lisis adaptado a predominio masculino")
print("   üìä Features espec√≠ficas de riesgo pulmonar")
print("   üó∫Ô∏è Clustering adaptado a patrones geogr√°ficos")
print("   üìà Proyecciones con factores espec√≠ficos")

print(f"\nüöÄ ¬°EJECUTA main() PARA COMENZAR!")

# ============================================================================
# NOTAS SOBRE ADAPTACIONES REALIZADAS
# ============================================================================

"""
üî¨ ADAPTACIONES ESPEC√çFICAS PARA C√ÅNCER DE PULM√ìN:

1. DEMOGR√ÅFICAS:
   - Rangos etarios ajustados (√©nfasis en 60+ a√±os)
   - An√°lisis de predominio masculino
   - Factores de riesgo por edad espec√≠ficos

2. EPIDEMIOL√ìGICAS:
   - Variables de riesgo adaptadas
   - Clustering con features espec√≠ficas
   - An√°lisis temporal considerando patrones pulmonares

3. MACHINE LEARNING:
   - Features engineering espec√≠fico
   - Modelos con par√°metros ajustados
   - Validaci√≥n temporal robusta mantenida

4. VISUALIZACIONES:
   - Gr√°ficos con interpretaci√≥n espec√≠fica
   - √ânfasis en grupos de alto riesgo
   - Comparaciones relevantes para c√°ncer pulmonar

5. REPORTE:
   - Recomendaciones espec√≠ficas (cesaci√≥n tab√°quica, etc.)
   - Interpretaci√≥n epidemiol√≥gica adaptada
   - Comparaci√≥n con c√°ncer de mama

üéØ RESULTADO: Sistema completamente adaptado manteniendo
              la robustez del sistema original pero con
              especificidad para c√°ncer de pulm√≥n.
"""
````

````
def analyze_lung_cancer():
    """An√°lisis completo de c√°ncer de pulm√≥n simplificado"""

    print("\nü´Å INICIANDO AN√ÅLISIS DE C√ÅNCER DE PULM√ìN")
    print("=" * 60)

    # 1. CARGAR DATOS
    try:
        df = pd.read_csv('DATA_CAPULMON_2012a2023 (1).csv', encoding='cp1252')
        print(f"‚úÖ Dataset cargado: {len(df):,} registros √ó {df.shape[1]} columnas")

        # Proteger datos
        protection_system.register_data('dataset_pulmon', df)

    except Exception as e:
        print(f"‚ùå Error cargando datos: {e}")
        return None

    # 2. AN√ÅLISIS B√ÅSICO
    print(f"\nüìä AN√ÅLISIS EPIDEMIOL√ìGICO:")
    print(f"   Per√≠odo: {df['ANIO_REGIS'].min()}-{df['ANIO_REGIS'].max()}")
    print(f"   Estados: {df['ENTIDAD'].nunique()}")
    print(f"   Edad promedio: {df['EDAD'].mean():.1f} a√±os")

    # 3. DISTRIBUCI√ìN POR SEXO (CLAVE PARA C√ÅNCER PULMONAR)
    print(f"\nüë• DISTRIBUCI√ìN POR SEXO:")
    sexo_dist = df['SEXO'].value_counts()
    for sexo, count in sexo_dist.items():
        pct = count / len(df) * 100
        print(f"   {sexo}: {count:,} ({pct:.1f}%)")

    # Verificar predominio masculino
    if len(sexo_dist) >= 2:
        ratio = sexo_dist.iloc[0] / sexo_dist.iloc[1]
        print(f"   Ratio: {ratio:.2f}:1")

        # Interpretaci√≥n epidemiol√≥gica
        if sexo_dist.index[0] in ['Hombre', 'HOMBRE', 'M', 'Masculino']:
            print(f"   ‚úÖ Confirma patr√≥n t√≠pico: predominio masculino en c√°ncer pulmonar")
        else:
            print(f"   üìä Patr√≥n observado diferente al t√≠pico")

    # 4. AN√ÅLISIS ETARIO ESPEC√çFICO
    print(f"\nüìÖ AN√ÅLISIS ETARIO (C√ÅNCER PULMONAR):")

    # Casos en grupos de alto riesgo
    casos_60_mas = (df['EDAD'] >= 60).sum()
    pct_60_mas = casos_60_mas / len(df) * 100
    print(f"   Casos ‚â•60 a√±os: {casos_60_mas:,} ({pct_60_mas:.1f}%)")

    casos_70_mas = (df['EDAD'] >= 70).sum()
    pct_70_mas = casos_70_mas / len(df) * 100
    print(f"   Casos ‚â•70 a√±os: {casos_70_mas:,} ({pct_70_mas:.1f}%)")

    # 5. TOP ESTADOS
    print(f"\nüèÜ TOP 10 ESTADOS - C√ÅNCER DE PULM√ìN:")
    top_estados = df['ENTIDAD'].value_counts().head(10)
    for i, (estado, casos) in enumerate(top_estados.items(), 1):
        pct = casos / len(df) * 100
        print(f"   {i:2d}. {estado}: {casos:,} ({pct:.1f}%)")

    # 6. EVOLUCI√ìN TEMPORAL
    print(f"\nüìà EVOLUCI√ìN TEMPORAL:")
    casos_anuales = df['ANIO_REGIS'].value_counts().sort_index()

    if len(casos_anuales) >= 2:
        primer_a√±o = casos_anuales.index[0]
        ultimo_a√±o = casos_anuales.index[-1]
        primer_casos = casos_anuales.iloc[0]
        ultimo_casos = casos_anuales.iloc[-1]

        cambio_absoluto = ultimo_casos - primer_casos
        cambio_pct = (cambio_absoluto / primer_casos) * 100 if primer_casos > 0 else 0

        print(f"   {primer_a√±o}: {primer_casos:,} casos")
        print(f"   {ultimo_a√±o}: {ultimo_casos:,} casos")
        print(f"   Cambio: {cambio_absoluto:+,} casos ({cambio_pct:+.1f}%)")

    # 7. CREAR VISUALIZACI√ìN B√ÅSICA
    print(f"\nüìä CREANDO VISUALIZACIONES...")

    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('ü´Å An√°lisis Epidemiol√≥gico - C√°ncer de Pulm√≥n M√©xico', fontsize=16, fontweight='bold')

    # Gr√°fico 1: Evoluci√≥n temporal
    ax1 = axes[0, 0]
    casos_anuales.plot(kind='line', marker='o', ax=ax1, color='darkred', linewidth=3)
    ax1.set_title('üìà Evoluci√≥n Temporal')
    ax1.set_xlabel('A√±o')
    ax1.set_ylabel('Casos')
    ax1.grid(True, alpha=0.3)

    # Gr√°fico 2: Distribuci√≥n por edad
    ax2 = axes[0, 1]
    df['EDAD'].hist(bins=25, ax=ax2, color='lightcoral', alpha=0.7, edgecolor='black')
    ax2.axvline(df['EDAD'].mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {df["EDAD"].mean():.1f}')
    ax2.axvline(60, color='orange', linestyle=':', linewidth=2, label='Edad de riesgo (60+)')
    ax2.set_title('üë• Distribuci√≥n por Edad')
    ax2.set_xlabel('Edad')
    ax2.set_ylabel('Frecuencia')
    ax2.legend()

    # Gr√°fico 3: Top estados
    ax3 = axes[1, 0]
    top_estados_plot = df['ENTIDAD'].value_counts().head(8)
    top_estados_plot.plot(kind='barh', ax=ax3, color='darkgreen')
    ax3.set_title('üó∫Ô∏è Top 8 Estados')
    ax3.set_xlabel('Casos')

    # Gr√°fico 4: Distribuci√≥n por sexo
    ax4 = axes[1, 1]
    sexo_counts = df['SEXO'].value_counts()
    colors = ['lightblue', 'lightpink'] if len(sexo_counts) == 2 else ['lightblue', 'lightpink', 'lightgreen']
    sexo_counts.plot(kind='pie', ax=ax4, autopct='%1.1f%%', colors=colors[:len(sexo_counts)])
    ax4.set_title('üë• Distribuci√≥n por Sexo')
    ax4.set_ylabel('')

    plt.tight_layout()
    plt.show()

    print(f"‚úÖ An√°lisis b√°sico completado")

    # 8. MACHINE LEARNING B√ÅSICO
    print(f"\nü§ñ AN√ÅLISIS ML B√ÅSICO:")

    # Preparar datos por estado
    state_features = df.groupby('ENTIDAD').agg({
        'EDAD': ['mean', 'count'],
        'ANIO_REGIS': ['min', 'max']
    }).round(2)

    state_features.columns = ['edad_media', 'total_casos', 'a√±o_min', 'a√±o_max']
    state_features = state_features.reset_index()

    # Variables derivadas
    state_features['a√±os_activos'] = state_features['a√±o_max'] - state_features['a√±o_min'] + 1
    state_features['casos_por_a√±o'] = state_features['total_casos'] / state_features['a√±os_activos']

    # Clustering b√°sico
    from sklearn.cluster import KMeans
    X = state_features[['casos_por_a√±o', 'edad_media']].fillna(0)

    # Encontrar clusters √≥ptimos
    silhouette_scores = {}
    for k in range(2, 6):
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(X)
        if len(set(labels)) > 1:
            silhouette_scores[k] = silhouette_score(X, labels)

    if silhouette_scores:
        best_k = max(silhouette_scores.keys(), key=lambda k: silhouette_scores[k])

        # Clustering final
        kmeans_final = KMeans(n_clusters=best_k, random_state=42)
        state_features['cluster'] = kmeans_final.fit_predict(X)

        print(f"üó∫Ô∏è Clustering geogr√°fico: {best_k} grupos identificados")

        # Mostrar clusters
        for cluster_id in range(best_k):
            cluster_states = state_features[state_features['cluster'] == cluster_id]
            print(f"\nüìä Cluster {cluster_id}:")
            print(f"   Estados: {len(cluster_states)}")
            print(f"   Casos promedio: {cluster_states['total_casos'].mean():.1f}")
            print(f"   Edad promedio: {cluster_states['edad_media'].mean():.1f}")

            # Mostrar algunos estados del cluster
            for _, row in cluster_states.head(3).iterrows():
                print(f"   ‚Ä¢ {row['ENTIDAD']}: {row['total_casos']} casos")

    # Guardar resultados
    protection_system.register_data('analisis_final', {
        'dataset': df,
        'state_features': state_features,
        'silhouette_scores': silhouette_scores
    })

    protection_system.force_save()

    print(f"\nüéâ ¬°AN√ÅLISIS COMPLETADO EXITOSAMENTE!")
    print(f"ü´Å C√°ncer de pulm√≥n analizado: {len(df):,} casos")
    print(f"üõ°Ô∏è Datos protegidos autom√°ticamente")

    return df

# Ejecutar an√°lisis
analyze_lung_cancer()
````

````
# AN√ÅLISIS AVANZADO - Ejecuta esto para profundizar
def advanced_lung_cancer_analysis():
    """An√°lisis avanzado con Machine Learning y proyecciones"""

    print("ü§ñ AN√ÅLISIS AVANZADO DE C√ÅNCER DE PULM√ìN")
    print("=" * 50)

    # Cargar datos
    df = pd.read_csv('DATA_CAPULMON_2012a2023 (1).csv', encoding='cp1252')

    # 1. VALIDACI√ìN TEMPORAL (2012-2020 vs 2021-2023)
    train_data = df[df['ANIO_REGIS'] <= 2020]
    test_data = df[df['ANIO_REGIS'] >= 2021]

    print(f"üìä Validaci√≥n temporal:")
    print(f"   Entrenamiento: {len(train_data):,} casos (2012-2020)")
    print(f"   Validaci√≥n: {len(test_data):,} casos (2021-2023)")

    # 2. PROYECCIONES FUTURAS 2024-2030
    casos_anuales = df['ANIO_REGIS'].value_counts().sort_index()

    # Modelo de regresi√≥n lineal
    from sklearn.linear_model import LinearRegression
    X = casos_anuales.index.values.reshape(-1, 1)
    y = casos_anuales.values

    model = LinearRegression()
    model.fit(X, y)

    # Proyecciones
    future_years = np.array(range(2024, 2031)).reshape(-1, 1)
    future_cases = model.predict(future_years)

    print(f"\nüîÆ PROYECCIONES C√ÅNCER PULMONAR 2024-2030:")
    for year, cases in zip(range(2024, 2031), future_cases):
        print(f"   {year}: {int(cases):,} casos estimados")

    # 3. AN√ÅLISIS POR SEXO Y EDAD COMBINADO
    print(f"\nüë• AN√ÅLISIS DEMOGR√ÅFICO DETALLADO:")

    # Distribuci√≥n por sexo y grupos etarios
    df['GRUPO_EDAD'] = pd.cut(df['EDAD'],
                             bins=[0, 50, 60, 70, 80, 120],
                             labels=['<50', '50-59', '60-69', '70-79', '80+'])

    sexo_edad = df.groupby(['SEXO', 'GRUPO_EDAD']).size().unstack(fill_value=0)
    print("Distribuci√≥n por sexo y edad:")
    print(sexo_edad)

    # 4. FACTORES DE RIESGO GEOGR√ÅFICO
    print(f"\nüó∫Ô∏è AN√ÅLISIS DE RIESGO GEOGR√ÅFICO:")

    # Calcular tasa por 100,000 habitantes
    state_risk = df.groupby('ENTIDAD').agg({
        'EDAD': ['count', 'mean'],
        'POBLACION': 'first'
    }).round(2)

    state_risk.columns = ['casos', 'edad_media', 'poblacion']
    state_risk['tasa_100k'] = (state_risk['casos'] / state_risk['poblacion'] * 100000).round(2)
    state_risk = state_risk.sort_values('tasa_100k', ascending=False)

    print("Estados con mayor tasa por 100,000 habitantes:")
    print(state_risk.head(10)[['casos', 'tasa_100k', 'edad_media']])

    # 5. TENDENCIAS TEMPORALES POR ESTADO
    print(f"\nüìà TENDENCIAS TEMPORALES:")

    # Estados con mayor crecimiento
    top_states = df['ENTIDAD'].value_counts().head(5).index

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.suptitle('üìà Tendencias Temporales por Estado - C√°ncer de Pulm√≥n', fontsize=16)

    for i, state in enumerate(top_states):
        if i < 5:  # Solo 5 gr√°ficos
            ax = axes[i//3, i%3]
            state_data = df[df['ENTIDAD'] == state]
            yearly_cases = state_data['ANIO_REGIS'].value_counts().sort_index()

            yearly_cases.plot(kind='line', marker='o', ax=ax, color='darkred', linewidth=2)
            ax.set_title(f'{state}', fontweight='bold')
            ax.set_xlabel('A√±o')
            ax.set_ylabel('Casos')
            ax.grid(True, alpha=0.3)

    # Gr√°fico resumen en el √∫ltimo subplot
    ax = axes[1, 2]
    casos_anuales.plot(kind='bar', ax=ax, color='steelblue', alpha=0.7)
    ax.set_title('Total Nacional', fontweight='bold')
    ax.set_xlabel('A√±o')
    ax.set_ylabel('Casos')
    ax.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

    print(f"\n‚úÖ AN√ÅLISIS AVANZADO COMPLETADO")
    return df

# Ejecutar an√°lisis avanzado
advanced_lung_cancer_analysis()
````

````
# DASHBOARD EJECUTIVO COMPLETO - MORTALIDAD C√ÅNCER DE PULM√ìN
def create_executive_dashboard():
    """Crear dashboard ejecutivo completo y funcional"""

    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    import pandas as pd

    print("üìä CREANDO DASHBOARD EJECUTIVO COMPLETO...")

    # Cargar datos
    df = pd.read_csv('DATA_CAPULMON_2012a2023 (1).csv', encoding='cp1252')

    # Configurar estilo
    plt.style.use('default')
    sns.set_palette("viridis")

    # Crear figura principal
    fig = plt.figure(figsize=(24, 16))
    fig.suptitle('ü´Å DASHBOARD EJECUTIVO: MORTALIDAD POR C√ÅNCER DE PULM√ìN M√âXICO 2012-2023\n' +
                 'Sistema de An√°lisis Epidemiol√≥gico con Machine Learning',
                 fontsize=20, fontweight='bold', y=0.97)

    # 1. Evoluci√≥n temporal (Top izquierda)
    ax1 = plt.subplot(3, 4, 1)
    casos_anuales = df['ANIO_REGIS'].value_counts().sort_index()
    ax1.plot(casos_anuales.index, casos_anuales.values, 'o-', linewidth=4,
             markersize=10, color='darkred', markerfacecolor='red')

    # L√≠nea de tendencia
    z = np.polyfit(casos_anuales.index, casos_anuales.values, 1)
    p = np.poly1d(z)
    ax1.plot(casos_anuales.index, p(casos_anuales.index), "--",
             alpha=0.8, color='blue', linewidth=3, label='Tendencia')

    ax1.set_title('üìà EVOLUCI√ìN TEMPORAL\nDefunciones por A√±o', fontweight='bold', fontsize=14)
    ax1.set_xlabel('A√±o', fontsize=12)
    ax1.set_ylabel('Defunciones', fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    # Destacar a√±os clave
    a√±o_max = casos_anuales.idxmax()
    casos_max = casos_anuales.max()
    ax1.annotate(f'Pico: {a√±o_max}\n{casos_max:,} casos',
                xy=(a√±o_max, casos_max), xytext=(a√±o_max-1, casos_max+200),
                arrowprops=dict(arrowstyle='->', color='red', lw=2),
                fontsize=10, fontweight='bold', color='red')

    # 2. Distribuci√≥n por edad (Top centro-izquierda)
    ax2 = plt.subplot(3, 4, 2)
    edades = df['EDAD'].dropna()
    ax2.hist(edades, bins=30, alpha=0.8, color='lightcoral', edgecolor='black', linewidth=1)

    # L√≠neas de referencia
    ax2.axvline(edades.mean(), color='red', linestyle='--', linewidth=3,
                label=f'Media: {edades.mean():.1f} a√±os')
    ax2.axvline(60, color='orange', linestyle=':', linewidth=3, label='Edad riesgo: 60+')
    ax2.axvline(70, color='purple', linestyle=':', linewidth=3, label='Alto riesgo: 70+')

    ax2.set_title('üë• DISTRIBUCI√ìN POR EDAD\nal Momento del Fallecimiento', fontweight='bold', fontsize=14)
    ax2.set_xlabel('Edad (a√±os)', fontsize=12)
    ax2.set_ylabel('Frecuencia', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # 3. Top 10 Estados (Top centro-derecha)
    ax3 = plt.subplot(3, 4, 3)
    top_estados = df['ENTIDAD'].value_counts().head(10)

    # Colores degradados
    colors = plt.cm.Reds(np.linspace(0.4, 1, len(top_estados)))
    bars = ax3.barh(range(len(top_estados)), top_estados.values, color=colors)

    ax3.set_yticks(range(len(top_estados)))
    ax3.set_yticklabels([estado[:12] + '...' if len(estado) > 12 else estado
                         for estado in top_estados.index], fontsize=10)
    ax3.set_title('üó∫Ô∏è TOP 10 ESTADOS\nMayor Mortalidad Absoluta', fontweight='bold', fontsize=14)
    ax3.set_xlabel('Defunciones', fontsize=12)

    # Agregar valores en barras
    for i, bar in enumerate(bars):
        width = bar.get_width()
        ax3.text(width + width*0.01, bar.get_y() + bar.get_height()/2,
                f'{int(width):,}', ha='left', va='center', fontweight='bold', fontsize=9)

    # 4. Distribuci√≥n por sexo (Top derecha)
    ax4 = plt.subplot(3, 4, 4)
    sexo_counts = df['SEXO'].value_counts()

    # Colores espec√≠ficos por sexo
    colores_sexo = {'Hombre': 'lightblue', 'Mujer': 'lightpink'}
    colors = [colores_sexo.get(sexo, 'lightgray') for sexo in sexo_counts.index]

    wedges, texts, autotexts = ax4.pie(sexo_counts.values, labels=sexo_counts.index,
                                       autopct='%1.1f%%', colors=colors, startangle=90,
                                       textprops={'fontsize': 12, 'fontweight': 'bold'})

    ax4.set_title('üë• DISTRIBUCI√ìN POR SEXO\nPatr√≥n de Mortalidad', fontweight='bold', fontsize=14)

    # Agregar ratio
    if len(sexo_counts) >= 2:
        ratio = sexo_counts.iloc[0] / sexo_counts.iloc[1]
        ax4.text(0, -1.3, f'Ratio: {ratio:.1f}:1', ha='center', fontsize=12,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7),
                fontweight='bold')

    # 5. Grupos etarios espec√≠ficos (Medio izquierda)
    ax5 = plt.subplot(3, 4, 5)
    grupos_edad = pd.cut(df['EDAD'], bins=[0, 50, 60, 70, 80, 120],
                        labels=['<50', '50-59', '60-69', '70-79', '80+'])
    grupo_counts = grupos_edad.value_counts().sort_index()

    # Colores por riesgo
    colores_riesgo = ['lightgreen', 'yellow', 'orange', 'red', 'darkred']
    bars = ax5.bar(range(len(grupo_counts)), grupo_counts.values,
                   color=colores_riesgo[:len(grupo_counts)], alpha=0.8, edgecolor='black')

    ax5.set_xticks(range(len(grupo_counts)))
    ax5.set_xticklabels(grupo_counts.index, fontsize=11)
    ax5.set_title('üìä GRUPOS ETARIOS\nDistribuci√≥n de Mortalidad', fontweight='bold', fontsize=14)
    ax5.set_ylabel('Defunciones', fontsize=12)

    # Agregar valores en barras
    for bar in bars:
        height = bar.get_height()
        ax5.text(bar.get_x() + bar.get_width()/2, height + height*0.01,
                f'{int(height):,}', ha='center', va='bottom', fontweight='bold', fontsize=10)

    # 6. Tendencias por estados principales (Medio centro-izquierda)
    ax6 = plt.subplot(3, 4, 6)
    estados_principales = ['Ciudad de M√©xico', 'M√©xico', 'Jalisco', 'Baja California', 'Chihuahua']
    colors_estados = plt.cm.Set1(np.linspace(0, 1, len(estados_principales)))

    for i, estado in enumerate(estados_principales):
        if estado in df['ENTIDAD'].values:
            estado_data = df[df['ENTIDAD'] == estado]
            casos_anuales_estado = estado_data['ANIO_REGIS'].value_counts().sort_index()
            if len(casos_anuales_estado) > 2:
                ax6.plot(casos_anuales_estado.index, casos_anuales_estado.values, 'o-',
                        label=estado[:10], color=colors_estados[i], linewidth=3, markersize=6)

    ax6.set_title('üìà TENDENCIAS REGIONALES\nTop 5 Estados', fontweight='bold', fontsize=14)
    ax6.set_xlabel('A√±o', fontsize=12)
    ax6.set_ylabel('Defunciones', fontsize=12)
    ax6.legend(fontsize=10)
    ax6.grid(True, alpha=0.3)

    # 7. An√°lisis de concentraci√≥n (Medio centro-derecha)
    ax7 = plt.subplot(3, 4, 7)
    estado_counts = df['ENTIDAD'].value_counts()
    cumsum_pct = (estado_counts.cumsum() / estado_counts.sum() * 100)

    # Gr√°fico de Pareto
    bars = ax7.bar(range(min(12, len(estado_counts))),
                   estado_counts.head(12).values, alpha=0.7, color='lightblue')

    ax7_twin = ax7.twinx()
    ax7_twin.plot(range(min(12, len(cumsum_pct))), cumsum_pct.head(12).values,
                  'ro-', color='red', linewidth=3, markersize=8)
    ax7_twin.set_ylabel('% Acumulado', color='red', fontsize=12, fontweight='bold')
    ax7_twin.tick_params(axis='y', labelcolor='red')

    # L√≠nea del 80%
    ax7_twin.axhline(y=80, color='red', linestyle='--', alpha=0.7, linewidth=2)
    ax7_twin.text(8, 82, '80% Regla Pareto', color='red', fontweight='bold')

    ax7.set_title('üìä CONCENTRACI√ìN GEOGR√ÅFICA\nAn√°lisis 80/20', fontweight='bold', fontsize=14)
    ax7.set_xlabel('Estados (Top 12)', fontsize=12)
    ax7.set_ylabel('Defunciones', fontsize=12)

    # 8. Resumen de m√©tricas clave (Medio derecha)
    ax8 = plt.subplot(3, 4, 8)

    # Calcular m√©tricas
    total_casos = len(df)
    edad_promedio = df['EDAD'].mean()
    casos_70_mas = (df['EDAD'] >= 70).sum()
    estados_count = df['ENTIDAD'].nunique()
    a√±os_estudio = df['ANIO_REGIS'].nunique()

    metricas = ['Total\nDefunciones', 'Estados\nAnalizados', 'A√±os\nEstudio',
                'Edad\nPromedio', 'Casos\n‚â•70 a√±os']
    valores = [total_casos/1000, estados_count, a√±os_estudio, edad_promedio, casos_70_mas/1000]

    colors_metricas = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    bars = ax8.bar(metricas, valores, color=colors_metricas, alpha=0.8, edgecolor='black')

    ax8.set_title('üìà M√âTRICAS CLAVE\nResumen Ejecutivo', fontweight='bold', fontsize=14)
    ax8.set_ylabel('Valor', fontsize=12)

    # Agregar valores en barras
    for i, (bar, valor) in enumerate(zip(bars, valores)):
        if i in [0, 4]:  # Para totales en miles
            ax8.text(bar.get_x() + bar.get_width()/2, bar.get_height() + bar.get_height()*0.02,
                    f'{valor:.1f}K', ha='center', va='bottom', fontweight='bold', fontsize=11)
        elif i == 3:  # Para edad
            ax8.text(bar.get_x() + bar.get_width()/2, bar.get_height() + bar.get_height()*0.02,
                    f'{valor:.1f}', ha='center', va='bottom', fontweight='bold', fontsize=11)
        else:
            ax8.text(bar.get_x() + bar.get_width()/2, bar.get_height() + bar.get_height()*0.02,
                    f'{int(valor)}', ha='center', va='bottom', fontweight='bold', fontsize=11)

    # 9-12. An√°lisis adicionales (Fila inferior)

    # 9. Distribuci√≥n temporal detallada
    ax9 = plt.subplot(3, 4, 9)
    casos_mensuales = df.groupby(['ANIO_REGIS']).size()
    ax9.bar(casos_mensuales.index, casos_mensuales.values, alpha=0.7, color='steelblue')
    ax9.set_title('üìä DISTRIBUCI√ìN ANUAL\nDetalle Temporal', fontweight='bold', fontsize=14)
    ax9.set_xlabel('A√±o', fontsize=12)
    ax9.set_ylabel('Defunciones', fontsize=12)
    ax9.tick_params(axis='x', rotation=45)

    # 10. Comparaci√≥n quinquenios
    ax10 = plt.subplot(3, 4, 10)
    df['QUINQUENIO'] = pd.cut(df['ANIO_REGIS'],
                             bins=[2011, 2015, 2020, 2023],
                             labels=['2012-2015', '2016-2020', '2021-2023'])
    quinquenio_counts = df['QUINQUENIO'].value_counts().sort_index()

    bars = ax10.bar(range(len(quinquenio_counts)), quinquenio_counts.values,
                    color=['lightcoral', 'orange', 'lightgreen'], alpha=0.8)
    ax10.set_xticks(range(len(quinquenio_counts)))
    ax10.set_xticklabels(quinquenio_counts.index, rotation=0, fontsize=10)
    ax10.set_title('üìä COMPARACI√ìN\nPor Per√≠odos', fontweight='bold', fontsize=14)
    ax10.set_ylabel('Defunciones', fontsize=12)

    # 11. Perfil demogr√°fico combinado
    ax11 = plt.subplot(3, 4, 11)

    # Crear matriz de edad vs sexo
    if 'SEXO' in df.columns:
        sexo_edad_matrix = df.groupby(['SEXO', pd.cut(df['EDAD'], bins=5)]).size().unstack(fill_value=0)
        if not sexo_edad_matrix.empty:
            sexo_edad_matrix.plot(kind='bar', stacked=True, ax=ax11, alpha=0.8)
            ax11.set_title('üë• PERFIL DEMOGR√ÅFICO\nEdad √ó Sexo', fontweight='bold', fontsize=14)
            ax11.set_xlabel('Sexo', fontsize=12)
            ax11.set_ylabel('Defunciones', fontsize=12)
            ax11.legend(title='Grupos Edad', fontsize=8)
            ax11.tick_params(axis='x', rotation=0)

    # 12. Proyecci√≥n y conclusiones
    ax12 = plt.subplot(3, 4, 12)

    # Proyecci√≥n simple
    a√±os_futuros = np.array([2024, 2025, 2026, 2027, 2028])
    √∫ltimos_3_a√±os = casos_anuales.iloc[-3:].mean()
    proyecci√≥n = [√∫ltimos_3_a√±os] * len(a√±os_futuros)

    # Combinar hist√≥rico y proyecci√≥n
    a√±os_completos = np.concatenate([casos_anuales.index, a√±os_futuros])
    valores_completos = np.concatenate([casos_anuales.values, proyecci√≥n])

    # Separar para colores diferentes
    hist√≥rico_x = casos_anuales.index
    hist√≥rico_y = casos_anuales.values
    futuro_x = a√±os_futuros
    futuro_y = proyecci√≥n

    ax12.plot(hist√≥rico_x, hist√≥rico_y, 'o-', color='blue', linewidth=3,
              markersize=6, label='Hist√≥rico')
    ax12.plot(futuro_x, futuro_y, 'o--', color='red', linewidth=3,
              markersize=6, alpha=0.7, label='Proyecci√≥n')

    ax12.set_title('üîÆ PROYECCI√ìN 2024-2028\nTendencia Esperada', fontweight='bold', fontsize=14)
    ax12.set_xlabel('A√±o', fontsize=12)
    ax12.set_ylabel('Defunciones', fontsize=12)
    ax12.legend()
    ax12.grid(True, alpha=0.3)

    # Ajustar layout
    plt.tight_layout()

    # Guardar dashboard
    plt.savefig('Dashboard_Ejecutivo_Cancer_Pulmon.png', dpi=300, bbox_inches='tight',
                facecolor='white', edgecolor='none')

    # Mostrar dashboard
    plt.show()

    # Resumen final
    print(f"""
üéØ DASHBOARD EJECUTIVO GENERADO EXITOSAMENTE
‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê

üìä CONTENIDO DEL DASHBOARD:
   ‚úÖ 12 visualizaciones ejecutivas
   ‚úÖ An√°lisis temporal completo
   ‚úÖ Distribuci√≥n demogr√°fica
   ‚úÖ Concentraci√≥n geogr√°fica
   ‚úÖ Proyecciones futuras
   ‚úÖ M√©tricas clave resumidas

üíæ ARCHIVO GUARDADO:
   üìÅ Dashboard_Ejecutivo_Cancer_Pulmon.png
   üìè Resoluci√≥n: 300 DPI (alta calidad)
   üìä Tama√±o: 24√ó16 pulgadas
   üé® Formato: PNG (listo para presentaciones)

üéØ USO RECOMENDADO:
   üìã Presentaciones ejecutivas
   üìä Reportes de salud p√∫blica
   üìà Toma de decisiones estrat√©gicas
   üè• Planificaci√≥n de recursos sanitarios
""")

# Ejecutar dashboard completo
create_executive_dashboard()
````
